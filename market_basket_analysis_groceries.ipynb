```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
from mlxtend.preprocessing import TransactionEncoder # OR...from mlxtend.preprocessing import OnehotTransactions
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('/content/GroceryStoreDataSet.csv', header=None)
df.head(3)
```

```python
df.columns = ['items']
df.head(3)
```

```python
txn_arr = df['items'].values
txn_arr[0:10] # diplaying first 10 transactions in txn_arr
```

```python
txn_list = [] # creating an empty transactions list called txn_list
for i in range(0,len(txn_arr)):
    txn_list.append(txn_arr[i].split(',')) # split all transactions in txn_arr on "," and append to txn_list


txn_list[0:10] # displaying first 10 transactions in txn_list
```

```python
enc = TransactionEncoder()
txn_list_enc = enc.fit(txn_list).transform(txn_list)
txn_list_enc[0:10] # displaying first 10 transactions in encoded transactions list txn_list_enc
```

```python
txn_df_enc = pd.DataFrame(data = txn_list_enc,
                          columns = enc.columns_,
                          dtype = int)             # note that columns = enc.colmns_ and not txn_df_enc.columns_
                                                   # set dtype = int, or else Boolean values will appear => will have...
                                                   # ...to perform txn_df_enc.astype(int)
txn_df_enc.head(3)
```

* Support of X (Sx) = Nx/N
* Support of Y (Sy) = Ny/N
* Support of (X -> Y) = Nxy/N

```python
df_support = apriori(txn_df_enc, use_colnames = True, min_support = 0.2) # min_support --> minimum support value filtered results
df_support.sort_values(by = 'support', ascending = False)
```

* Lift (X -> Y) = Nxy/Nx*Ny

```python
df_lift = association_rules(df_support, metric = 'lift', min_threshold = 1) # displaying results with lift >= 1
df_lift.sort_values(by = 'lift', ascending = False)
```

Since these items have lift > 1, they are usually ................. products

```python
df_lift = association_rules(df_support, metric = 'lift', min_threshold = 0) # displaying results with lift >= 0
df_lift[df_lift['lift'] < 1] # displaying results with lift < 1
```

These items with lift < 1 are usually ............ of one another

```python
df_lift = association_rules(df_support, metric = 'lift', min_threshold = 0) # displaying results with lift >= 0
df_lift[df_lift['lift'] == 1] # displaying results with lift = 1
```

There are no items with lift = 1, it means that.............

* Confidence (X -> Y) = [P(Y | X) / P(X)] = [(Nxy/N) / (Nx/N)] = Nxy/Nx

```python
df_confidence = association_rules(df_support, metric = 'confidence', min_threshold = 0.3)
                                                                       # displaying confidence > 30%
df_confidence.sort_values(by = 'confidence', ascending = False)
```
